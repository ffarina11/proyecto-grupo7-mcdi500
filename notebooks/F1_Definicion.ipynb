## 1. Contextualización del proyecto

El rendimiento académico de los estudiantes es un fenómeno multidimensional influenciado por factores individuales, familiares, socioeconómicos e institucionales. En contextos de educación secundaria, la identificación de los determinantes del éxito o fracaso académico resulta fundamental para el diseño de políticas educativas efectivas y para la orientación de los recursos de apoyo estudiantil.

El presente proyecto analiza el **Student Performance Dataset** (Cortez & Silva, 2008), disponible en el UCI Machine Learning Repository. Este conjunto de datos fue recopilado en dos establecimientos educacionales de Portugal y contiene información sobre **649 estudiantes** de dos asignaturas —Matemáticas y Portugués— con **33 variables** que describen características demográficas, familiares, socioeconómicas, hábitos de estudio y calificaciones parciales y finales.

El proyecto se desarrolla en el marco del curso **MCDI500 — Herramientas de Software Científico** del Magíster en Ciencias de Datos e Inteligencia Artificial de la Universidad Andrés Bello, aplicando un flujo de trabajo reproducible mediante Python, Jupyter Lab, Git y GitHub.

## 2. Definición de la problemática

### 2.1 Problema de investigación

A pesar de que numerosos estudios han identificado variables asociadas al rendimiento académico, persiste la pregunta de **qué combinación de factores socioeconómicos y de preparación previa explica mejor las diferencias en las calificaciones finales** entre estudiantes de distintos establecimientos educacionales.

El problema central de este proyecto es:

> *¿Qué factores socioeconómicos y de preparación previa explican las diferencias en el rendimiento académico entre estudiantes de dos establecimientos educacionales, considerando dos asignaturas diferentes?*

### 2.2 Preguntas de investigación

1. ¿Qué variables (nivel educativo de los padres, tiempo de estudio, apoyo familiar, entre otras) se asocian con mayor fuerza al rendimiento académico medido por la calificación final (G3)?
2. ¿Existen diferencias estadísticamente relevantes en el rendimiento entre los dos establecimientos educacionales (`school: GP` y `school: MS`)?
3. ¿Qué peso tiene el contexto socioeconómico del estudiante (educación y ocupación de los padres, acceso a internet, recursos en el hogar) en la predicción del rendimiento final?

## 5. Descripción del dataset

El **Student Performance Dataset** (Cortez & Silva, 2008) contiene datos de estudiantes de educación secundaria de dos escuelas portuguesas. Sus características principales son:

| Característica | Detalle |
|---|---|
| Fuente | UCI Machine Learning Repository |
| Archivos | `student-mat.csv` (Matemáticas) · `student-por.csv` (Portugués) |

### Grupos de variables

| Grupo | Variables |
|---|---|
| Demográficas | `school`, `sex`, `age`, `address`, `famsize`, `Pstatus` |

| Hábitos y apoyo | `studytime`, `failures`, `schoolsup`, `famsup`, `paid`, `activities`, `nursery`, `higher`, `internet`, `romantic` |
| Contexto familiar | `famrel`, `freetime`, `goout`, `Dalc`, `Walc`, `health`, `absences` |
| Calificaciones | `G1`, `G2`, `G3` (notas 1°, 2° período y final) |

## 7. Carga inicial del dataset

Se cargan los dos archivos CSV del Student Performance Dataset desde la carpeta `data/raw/`. Los archivos deben estar previamente descargados desde el [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/320/student+performance) y ubicados en esa carpeta del repositorio.

**Decisión técnica:** se cargan ambos datasets por separado para preservar la integridad de los datos originales, conforme al principio de reproducibilidad que establece que los datos crudos (`raw`) no deben modificarse.

````python
# ============================================================
# CARGA DEL DATASET
# Los archivos CSV usan punto y coma (;) como separador,
# por lo que se especifica sep=';' en la lectura.
# ============================================================

# Rutas relativas desde la raíz del repositorio
RUTA_MAT = "../data/raw/student-mat.csv"
RUTA_POR = "../data/raw/student-por.csv"

# Carga de datos
df_mat = pd.read_csv(RUTA_MAT, sep=";")
df_por = pd.read_csv(RUTA_POR, sep=";")

print("Dataset cargado exitosamente.")
print(f"  student-mat.csv : {df_mat.shape[0]} filas × {df_mat.shape[1]} columnas")
print(f"  student-por.csv : {df_por.shape[0]} filas × {df_por.shape[1]} columnas")
````
