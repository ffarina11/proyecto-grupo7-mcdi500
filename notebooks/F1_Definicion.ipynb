## 1. Contextualización del proyecto

El rendimiento académico de los estudiantes es un fenómeno multidimensional influenciado por factores individuales, familiares, socioeconómicos e institucionales. En contextos de educación secundaria, la identificación de los determinantes del éxito o fracaso académico resulta fundamental para el diseño de políticas educativas efectivas y para la orientación de los recursos de apoyo estudiantil.

El presente proyecto analiza el **Student Performance Dataset** (Cortez & Silva, 2008), disponible en el UCI Machine Learning Repository. Este conjunto de datos fue recopilado en dos establecimientos educacionales de Portugal y contiene información sobre **649 estudiantes** de dos asignaturas —Matemáticas y Portugués— con **33 variables** que describen características demográficas, familiares, socioeconómicas, hábitos de estudio y calificaciones parciales y finales.

El proyecto se desarrolla en el marco del curso **MCDI500 — Programación para la Ciencia de Datos** del Magíster en Ciencias de Datos e Inteligencia Artificial de la Universidad Andrés Bello, aplicando un flujo de trabajo reproducible mediante Python, Jupyter Lab, Git y GitHub.

## 2. Definición de la problemática

### 2.1 Problema de investigación

A pesar de que numerosos estudios han identificado variables asociadas al rendimiento académico, es un fenómeno complejo influenciado por múltiples factores de carácter individual, familiar, socioeconómico y educativo. En este contexto, la problemática abordada en el presente proyecto consiste en **determinar qué factores socioeconómicos y de preparación previa podrían explicar las diferencias en el rendimiento académico entre estudiantes de dos establecimientos educacionales.**

### 2.2 Preguntas de investigación
Con el fin de orientar el desarollo del proyecto y la búsqueda de evidencias, métricas y patrones relevantes se plantan 5 preguntas de investigación:

1. ¿Qué factores socioeconómicos y familiares presentan una mayor asociación con el rendimiento académico de los estudiantes?
2. ¿De qué manera las características individuales y los hábitos de estudio influyen en las calificaciones obtenidas?
3. ¿Existen diferencias significativas en el rendimiento académico entre estudiantes de distintos establecimientos educacionales?
4. ¿Qué variables presentan una mayor capacidad explicativa respecto de las diferencias observadas en los resultados académicos?
5. ¿Qué variables presentan una mayor capacidad predictiva sobre la calificación final G3?

<a id='5-dataset'></a>
## 5. Descripción del dataset

El **Student Performance Dataset** (Cortez & Silva, 2008) contiene datos de estudiantes de educación secundaria de dos escuelas portuguesas. Sus características principales son:

| Característica | Detalle |
|---|---|
| Fuente | UCI Machine Learning Repository |
| Archivos | `student-mat.csv` (Matemáticas) · `student-por.csv` (Portugués) |
| Filas | 395 (mat) + 649 (por) |
| Columnas | 33 variables por archivo |
| Variable objetivo | `G3` — calificación final (0–20) |

### Grupos de variables

| Grupo | Variables |
|---|---|
| Demográficas | `school`, `sex`, `age`, `address`, `famsize`, `Pstatus` |
| Socioeconómicas | `Medu`, `Fedu`, `Mjob`, `Fjob`, `reason`, `guardian` |
| Hábitos y apoyo | `studytime`, `failures`, `schoolsup`, `famsup`, `paid`, `activities`, `nursery`, `higher`, `internet`, `romantic` |
| Contexto familiar | `famrel`, `freetime`, `goout`, `Dalc`, `Walc`, `health`, `absences` |
| Calificaciones | `G1`, `G2`, `G3` (notas 1°, 2° período y final) |



### Descripción de Variables 

| Nombre de la variable | Tipo | Descripción |
| :--- | :--- | :--- |
| **school** | Categórico | Escuela del estudiante (binario: 'GP' - Gabriel Pereira o 'MS' - Mousinho da Silveira) |
| **sex** | Binario | Sexo del estudiante (binario: 'F' - femenino o 'M' - masculino) |
| **age** | Numérico | Edad del estudiante (numérica: de 15 a 22 años) |
| **address** | Categórico | Tipo de dirección de domicilio del estudiante (binario: 'U' - urbano o 'R' - rural) |
| **famsize** | Categórico | Tamaño de familia (binario: 'LE3' - menor o igual a 3 o 'GT3' - mayor que 3) |
| **Pstatus** | Categórico | Estado de convivencia de los padres (binario: 'T' - vivir juntos o 'A' - separados) |
| **Medu** | Numérico | Educación materna (numérica: 0 - ninguna, 1 - educación primaria (4º curso), 2 - de 5º a 9º curso, 3 - educación secundaria o 4 - educación superior) |
| **Fedu** | Numérico | Educación del padre (numérica: 0 - ninguno, 1 - educación primaria (4º curso), 2 - de 5º a 9º curso, 3 - educación secundaria o 4 - educación superior) |
| **Mjob** | Categórico | Trabajo de la madre (nominal: 'teacher', 'health', 'services' (civil), 'at_home' u 'other') |
| **Fjob** | Categórico | Trabajo del padre (nominal: 'teacher', 'health', 'services' (civil), 'at_home' u 'other') |
| **reason** | Categórico | Motivo para elegir esta escuela (nominal: cerca de 'home', 'reputation', 'course' u 'other') |
| **guardian** | Categórico | Tutor del estudiante (nominal: 'mother', 'father' u 'other') |
| **traveltime** | Numérico | Tiempo de viaje de casa a la escuela (numérico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, o 4 - >1 hora) |
| **studytime** | Numérico | Tiempo de estudio semanal (numérico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas, o 4 - >10 horas) |
| **failures** | Numérico | Número de fracasos de clase pasados (numérico: n si 1<=n<3, si no 4) |
| **schoolsup** | Binario | Apoyo educativo extra (binario: sí o no) |
| **famsup** | Binario | Apoyo educativo familiar (binario: sí o no) |
| **paid** | Binario | Clases extra de pago dentro de la asignatura (Matemáticas o Portugués) (binario: sí o no) |
| **activities** | Binario | Actividades extraescolares (binario: sí o no) |
| **nursery** | Binario | Asistió a una guardería (binario: sí o no) |
| **higher** | Binario | Quiere cursar estudios superiores (binario: sí o no) |
| **internet** | Binario | Acceso a Internet en casa (binario: sí o no) |
| **romantic** | Binario | Con una relación romántica (binario: sí o no) |
| **famrel** | Numérico | Calidad de las relaciones familiares (numéricas: de 1 - muy mala a 5 - excelente) |
| **freetime** | Numérico | Tiempo libre después del colegio (numérico: de 1 - muy bajo a 5 - muy alto) |
| **g1** | Categórico | Nota del primer periodo (numérica: de 0 a 20) |
| **g2** | Categórico | Nota del segundo periodo (numérica: de 0 a 20) |
| **g3** | Numérico | Nota final (numérica: de 0 a 20, variable objetivo / target) |

Decisión Ténica: Durante la fase 1 se optó por mantener las dos fuentes de datos en archivos independientes y cargarlas por separado en Jupyter Lab, con el fin de preservar su estructura original y asegurar la trazabilidad y reproducibilidad del proceso. Esta decisión evita transformaciones prematuras que pudieran introducir errores o sesgos en los datos.
